```python
import numpy as np
from sklearn import kernel_approximation
```

```python
rng=np.random.RandomState(0)
X=rng.rand(10,2000)
X=np.array(X,dtype='float32')
print(X.dtype)
```

```python
transformer=kernel_approximation.RBFSampler()
X_new=transformer.fit_transform(X)
print(X_new.dtype)
```

At first, X is float32 and is unchanged by fit_transform(X) where it is still maintaing the datatype.

float-32 typed data is often more efficient than using float-64. Reason being, it reduce memory usage and processing time by leveraging vector instructions of CPU.

Be cautious because it can sometimes lead to numerical stability problems causing algorithm to be more sensitive to the scale of the values and require adequate processing

```python
from sklearn import datasets
from sklearn.svm import SVC
iris=datasets.load_iris()
```

```python
clf=SVC()
clf.fit(iris.data,iris.target)
list(clf.predict(iris.data[:3]))
```

```python
clf.fit(iris.data, iris.target_names[iris.target])
list(clf.predict(iris.data[:3]))
```

 the first predict() returns an integer array, since iris.target (an integer array) was used in fit. The second predict() returns a string array, since iris.target_names was for fitting.
