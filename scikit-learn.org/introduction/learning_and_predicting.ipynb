Given sample of each of the 10 possible classes (the digits zero through nine) on which we fit an estimator to able to predict the classes to which unseen samples belong

fit(X,y) and predict(T) is the methods we will use to fit an estimator for classification

```python
clf=svm.SVC(gamma=0.001,C=100)
# Here gamma values are set manually but to find good values for these parameters, use grid search and cross validation
```

```python
digits=datasets.load_digits()
clf.fit(digits.data[:-1],digits.target[:-1])
```

We select the training set with the [:-1] Python syntax, which produces a new array that contains all but the last item from digits.data

```python
clf.predict(digits.data[-1:])
```
